```python
v1 = np.array([0,1])
v2 = np.array([1,0])

plot_vectors([v1,v2], ["blue","red"])

plt.xlim(-0.25, 2)
plt.ylim(-0.25, 2)

plt.show()
```

```python
A = np.array([[2,0],[0,2]])
print(A)
```

```python
v1_transformed = A.dot(v1)
v2_transformed = A.dot(v2)

plot_vectors([v1_transformed,v2_transformed], ["orange","green"])

plt.xlim(-0.25, 3)
plt.ylim(-0.25, 3)

plt.show()
```

```python
det_A = np.linalg.det(A)

print(det_A)
```

```python
transformed_area = abs((v1_transformed[0]-v2_transformed[0]))*abs((v1_transformed[1]-v2_transformed[1]))
print(transformed_area)
```

```python
B = A * [-1, 1]
print(B)
```

```python
det_B = np.linalg.det(B)
print(det_B)
```

```python
v1 = np.array([0,1])
v2 = np.array([1,0])

v1_transformed = B.dot(v1)
v2_transformed = B.dot(v2)

plot_vectors([v1_transformed, v2_transformed, v1, v2],
                ['blue','red','green','orange'])

plt.xlim(-3, 1.5)
plt.ylim(-0.5, 2.5)

plt.show()
```
