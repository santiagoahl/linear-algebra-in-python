## Inverse of diagonal matrices

They have easy-to-compute matrices, in fact, the inverse of a diagonal matrix $D$ is a diagonal matrix which entries are the inverses of the original one

```python
diag_inv = np.diag(1/v)
```

```python
identity = diag_matrix.dot(diag_inv)
identity
```

#Symmetric matrix

Recall that, if $A = [a_{ij}]_{i,j}$, $B = [b_{ij}]_{i,j}$ are matrices, the transpose are defined by
$$A^T:= [a_{ji}]_{i,j}$$

We say that a matrix $S$ is symmetric iff $S=S^T$.

```python
S = np.array([
    [1,2,3],
    [2,-1,7],
    [3, 7,11]
])
S
```

```python
S.T
```

# Orthogonal vectors

Two vectors $\text{v_1,v_2}$ are orthogonal iff its angle is $\frac{\pi}{2}$, which is equivalent as $\text{v_1}\cdot\text{v_2}=0$.

They are orthonormal if else each one is normal. 

```python
x = np.array([0, 0, 2, 2])
y = np.array([0, 0, 2, -2])

plt.quiver([x[0], y[0]],
           [x[1], y[1]],
           [x[2], y[2]],
           [x[3], y[3]],
           angles='xy', scale_units='xy', scale=1)

plt.xlim(-2,4)
plt.ylim(-3,3)
plt.show()
```

```python
angle = np.arccos(x.dot(y.T)/(np.linalg.norm(x)*np.linalg.norm(y)))
angle
```

# Orthogonal matrix

A matrix $O$ is said *orthogonal* if its column vectors are 2-2 orthogonal.

If every column vector and every row vector is normal, we say that $O$ is orthonormal.

If $O$ is orthonormal, then:

$$
  A^{-1} = A^T
$$

```python
O = np.array([
    [1,0,0],
    [0,1,0],
    [0, 0,1]
])
O
```

```python
print(O[:,0].dot(O[:,1]))
print(O[:,0].dot(O[:,2]))
print(O[:,1].dot(O[:,2]))
```

```python
print(O[0,:].dot(O[1,:]))
print(O[0,:].dot(O[2,:]))
print(O[1,:].dot(O[2,:]))
```

```python
print(np.linalg.norm(O[0,:]))
print(np.linalg.norm(O[1,:]))
print(np.linalg.norm(O[2,:]))
```

## Example

```python
A = np.array([[np.cos(100), -np.sin(100)],
             [np.sin(100), np.cos(100)]])
print(A)
```

```python
print(np.linalg.norm(A[0,:]))
print(np.linalg.norm(A[1,:]))
print(np.linalg.norm(A[:,0]))
print(np.linalg.norm(A[:,1]))
```

```python
A_t = A.T
print(A_t.dot(A))
print(A.dot(A_t))
```

```python
D = A.dot(A_t)
```

```python
A_inv = np.linalg.inv(A)
print(A_inv)
```

```python
error = 1/A_t.dot(A)
print(error)
```

Trick to make approximations to zero.

It is about defining a very small value (epsilon) against which the elements of the matrix are compared and if they are less than it, then set them to zero:

```python
eps= 1e-16
D[np.abs(D) < eps] = 0
D
```
