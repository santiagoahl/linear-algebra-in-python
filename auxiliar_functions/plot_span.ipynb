```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_span(v1, v2, color):
  v1 = np.array(v1)
  v2 = np.array(v2)
  for lambda_1 in range(-10, 10):
    for lambda_2 in range(-10, 10):
      plt.scatter(lambda_1*v1[0]+lambda_2*v2[0], lambda_1*v1[1]+lambda_2*v2[1],
                  marker='.',
                  color=color)
  m = 10*int(abs(max(v1[0], v1[1])))+10*int(abs(max(v2[0], v2[1])))+10
  plt.xlim(-m, m)
  plt.ylim(-m, m)
  plt.axvline(x=0, color='grey')
  plt.axhline(y=0, color='grey')
  plt.show()
```
