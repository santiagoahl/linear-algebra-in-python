# Matrix diagonalization

Let $A$ be a matrix, if $A$ is diagonalizable, we can decompose $A$ as:

$$
  A = \begin{bmatrix}
\text{v}_1 \\
\text{v}_2 \\
\vdots \\
\text{v}_n
\end{bmatrix}
\begin{bmatrix}
\lambda_1 \\
&\lambda_2 \\
&&\ddots \\
&&&\lambda_n
\end{bmatrix}
\begin{bmatrix}
\text{v}_1 \\
\text{v}_2 \\
\vdots \\
\text{v}_n
\end{bmatrix}^{-1}
$$

Were $\{\text{v}_k\}_{k\leq n}$ are the eigenvectors of $A$ and $\{\lambda_k\}_{k\leq n}$ are the eigenvalues.

```python
A = np.array([
    [3, 2], 
    [4, 1]
])
eigenvalues, eigenvectors = np.linalg.eig(A)
```

```python
A_computed = eigenvectors.dot(np.diag(eigenvalues)).dot(np.linalg.inv(eigenvectors))
```

```python
print(A_computed)
```

If $A$ is symmetric and real valued, then

$$
  A = \begin{bmatrix}
\text{v}_1 \\
\text{v}_2 \\
\vdots \\
\text{v}_n
\end{bmatrix}
\begin{bmatrix}
\lambda_1 \\
&\lambda_2 \\
&&\ddots \\
&&&\lambda_n
\end{bmatrix}
\begin{bmatrix}
\text{v}_1 \\
\text{v}_2 \\
\vdots \\
\text{v}_n
\end{bmatrix}^{T}
$$

Here is easiest to decompose $A$ cuz computing the transpose is computationally lower than computing the inverse.

```python
A = np.array([
    [3, 2], 
    [2, 3]
])
eigenvalues, eigenvectors = np.linalg.eig(A)
```

```python
A == A.T
```

```python
print(eigenvalues, '\n\n',eigenvectors)
```

```python
A_computed = eigenvectors.dot(np.diag(eigenvalues)).dot(eigenvectors.T)
print(A_computed)
```

# Decomposition of non-square-matrices: SVD

If a matrix $A$ is not a square, then, we want to express $A$ as:

$$
  A = UVD
$$

Where $D$ is a diagonal matrix, and $U,V$ could be non-square matrices.

In mathematics, in particular functional analysis, the singular values, or s-numbers of a compact operator $T:X\rightarrow Y$  acting between Hilbert spaces $X$ and $Y$, are the square roots of the (necessarily non-negative) eigenvalues of the self-adjoint operator $T^{*}T$ (where $T^{*}T^{*}$ denotes the adjoint of $T$).

The singular values are non-negative real numbers, usually listed in decreasing order $(σ1(T), σ2(T), …)$. The largest singular value $σ1(T)$ is equal to the operator norm of T (see Min-max theorem).

```python
A = np.array([
    [1, 2, 3],
    [3, 4, 5]
])
print(A)
```

```python
U, D, V = np.linalg.svd(A)
```

```python
print(U)
```

```python
print(np.diag(D))
```

```python
print(V)
```
