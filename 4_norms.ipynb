# Norms

Norms are used to measure vectors. Are used in machine learning to measure the accurracy of our models.

## Definition

Given $V$ a vector space, we say that a function $$N:V→\mathbb{R}_{\geq0} $$
Is a norm iff

1. $N(v)\geq0$ iff $v=\vec{0}∀v\in V$.
2. $ N(v)=0$ iff $v=\vec{0}$.
3. $N(λv)=|λ|\cdot N(v), ∀v\in V, ∀\lambda\in \mathbb{R}$.
4. $N(v_1+v_2)\leq N(v_1)+N(v_2), ∀v_1,v_2\in V$ (Triangular innequality).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
v1 = np.array([2, 7])
v2 = np.array([3, 5])
```

```python
v1v2 = v1+v2
```

```python
normav1v2 = np.linalg.norm(v1v2)
print(normav1v2)
```

```python
normav1 = np.linalg.norm(v1)
normav2 = np.linalg.norm(v2)
print(normav1+normav2)
```

```python
normav1v2 <= normav1+normav2
```

## Visualize sum of vectors

```python
v1 = np.array([0,0,2,7])
v2 = np.array([0,0,3,5])

v1_aux = np.array([v1[2],v1[3],v2[2],v2[3]])
v1v2 = np.array([0,0,5,12])
```

```python
plt.quiver([v1[0], v1_aux[0], v1v2[0]],
           [v1[1], v1_aux[1], v1v2[1]],
           [v1[2], v1_aux[2], v1v2[2]],
           [v1[3], v1_aux[3], v1v2[3]],
          angles = 'xy', scale_units = 'xy', scale = 1,
           color = sns.color_palette()          
          )

plt.xlim(-0.5, 6)
plt.ylim(-0.5, 15)
```

# $L$-norms

Given $\text{v}=[v_1, v_2,\cdots,v_n]$ a vector, then

$L_0:= |\{v_i:v_i\neq0\}|$ (For example, count the number of users that are premium)

$L_1:= \sum_i|v_i|$ (For example, count the distance needed to move in Manhattan)

$L_2$:=Euclidean usual distance

$L_∞ := max_i|v_i|$

```python
v = np.array([1,2,0,5,6,0])
```

```python
l0 = np.linalg.norm(v, ord=0)
l0
```

```python
l1 = np.linalg.norm(v, ord=1)
l1
```

```python
l2 = np.linalg.norm(v, ord=2)
l2
```

```python
l_inf = np.linalg.norm(v, ord=np.inf)
l_inf
```

# Norm and Dot product

We have the following property:
$$
  v_1\cdot v_2 = ||v_1||_2||v_2||_2cos(\theta)
$$

Were $\theta$ is the angle between $v_1$ and $v_2$.

```python
v1 = np.array([0,0,0,3])
v2 = np.array([0,0,3,3])
```

```python
plt.xlim(-2,6)
plt.ylim(-2,6)

plt.quiver([v1[0], v2[0]],
           [v1[1], v2[1]],
           [v1[2], v2[2]],
           [v1[3], v2[3]],
          angles = 'xy', scale_units = 'xy', scale = 1,
           color = sns.color_palette()          
          )
plt.show()
```

```python
v1 = np.array([0,3])
v2 = np.array([3,3])

print(v1.T.dot(v2))
```

```python
nv1 = np.linalg.norm(v1)
nv2 = np.linalg.norm(v2)
cos = np.cos(np.deg2rad(45))
(nv1*nv2)*cos
```
